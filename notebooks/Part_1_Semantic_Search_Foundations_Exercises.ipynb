# Section 1: Foundation - Understanding Semantic Search

## 1A. Why Semantic Search Matters

Traditional keyword search requires exact matches. **Semantic search understands meaning.**

**Example:** "Something to keep my drinks cold"
- ❌ **Keyword:** Searches for exact words → often finds nothing  
- ✅ **Semantic:** Understands intent → returns insulated bottles, coolers, ice chests

### Keyword Search vs Semantic Search

<div style="background: linear-gradient(135deg, rgba(88,28,135,0.25) 0%, rgba(124,58,237,0.20) 100%);
            backdrop-filter: blur(8px);
            border: 1px solid rgba(124,58,237,0.4);

<div style="color: #ffffff;">

| Aspect | Keyword Search | Semantic Search |
|--------|----------------|-----------------|
| **What it is** | Exact text matching | Meaning-based matching |
| **How it works** | String comparison (LIKE, ILIKE) | Vector similarity (embeddings) |
| **Example query** | `WHERE name LIKE '%headphones%'` | `WHERE embedding <=> query < 0.3` |
| **User experience** | Must know exact product names | Natural language: "gift for music lover" |
| **Technology** | SQL text search | pgvector + Bedrock embeddings |

</div>

<p style="color: #ffffff; font-weight: 600; margin: 16px 0 8px 0;">
Bottom line: Keyword search finds what you type. Semantic search finds what you mean.
</p>

````python
# ============================================================================
# 📝 TODO: Interactive Product Insert Exercise
# ============================================================================

# Step 1: Personalize your product
# 💡 Look at the TOP of VS Code for input prompts (command palette area)
your_name = input("Enter your name: ")
product_description = input(f"Create a product description (e.g., '{your_name}'s Amazing Wireless Headphones'): ")

# Generate unique product ID: DAT + first 4 letters + 3 random digits = 10 chars
import random
name_parts = your_name.split()
first_four = name_parts[0][:4].upper().ljust(4, "X")
random_suffix = str(random.randint(100, 999))
product_id = f"DAT{first_four}{random_suffix}"

print(f"\n✅ Creating product: {product_description}")
print(f"📦 Product ID: {product_id}")

# Step 2: Generate embedding using Bedrock
print("\n🤖 Generating embedding with Bedrock Titan V2...")
response = bedrock.invoke_model(
    modelId='amazon.titan-embed-text-v2:0',
    body=json.dumps({"inputText": product_description})
)
embedding = json.loads(response['body'].read())['embedding']
print(f"✅ Generated {len(embedding)}-dimensional vector")

# Step 3: Insert into database
print("\n💾 Inserting into database...")
with psycopg.connect(conn_string, row_factory=dict_row) as conn:
    with conn.cursor() as cur:
        cur.execute("""
            INSERT INTO bedrock_integration.product_catalog 
            ("productId", product_description, category_name, price, stars, reviews, quantity, embedding)
            VALUES (%s, %s, %s, %s, %s, %s, %s, %s)
        """, (
            product_id,
            product_description,
            "Workshop Demo",
            99.99,
            5.0,
            1,
            100,
            str(embedding)
        ))
        conn.commit()

print(f"✅ Product inserted successfully!")

# Step 4: Search for your product using semantic search (<=> is the cosine distance operator)
print(f"\n🔍 Searching for products similar to: '{your_name}'...")
search_response = bedrock.invoke_model(
    modelId='amazon.titan-embed-text-v2:0',
    body=json.dumps({"inputText": your_name})
)
search_embedding = json.loads(search_response['body'].read())['embedding']

with psycopg.connect(conn_string, row_factory=dict_row) as conn:
    with conn.cursor() as cur:
        cur.execute("""
            SELECT 
                product_description,
                category_name,
                price,
                stars,
                1 - (embedding <=> %s::vector) as similarity
            FROM bedrock_integration.product_catalog
            WHERE 1 - (embedding <=> %s::vector) > 0.5
            ORDER BY embedding <=> %s::vector
            LIMIT 5
        """, (str(search_embedding), str(search_embedding), str(search_embedding)))
        
        results = cur.fetchall()

print(f"\n🎯 Found {len(results)} similar products:")
print("=" * 80)
for i, product in enumerate(results, 1):
    print(f"\n{i}. {product['product_description']}")
    print(f"   Category: {product['category_name']} | Price: ${product['price']:.2f}")
    print(f"   Rating: {'⭐' * int(product['stars'])} | Similarity: {product['similarity']:.3f}")

print("\n" + "=" * 80)
print("✅ SUCCESS! Your product is now in the database and searchable!")
print("\n🎨 Try searching for it in the Blaize Bazaar AI chat!")
````

````python
def generate_embedding(text: str) -> List[float]:
    """
    Generate semantic embeddings using Amazon Titan Text Embeddings V2.
    
    Converts natural language text into a 1024-dimensional vector representation
    that captures semantic meaning, enabling similarity-based search beyond
    keyword matching.
    
    Args:
        text: Input text to embed (query or document content)
        
    Returns:
        List of 1024 floating-point values representing the semantic embedding
    """
    response = bedrock.invoke_model(
        modelId="amazon.titan-embed-text-v2:0",
        body=json.dumps({"inputText": text})
    )
    
    result = json.loads(response['body'].read())
    return result['embedding']


# ============================================================================
# Embedding Generation Test
# ============================================================================

sample_query = "wireless headphones for running"
sample_embedding = generate_embedding(sample_query)

print("=" * 60)
print("✅ Embedding Generation Successful")
print("=" * 60)
print(f"Query Text:        '{sample_query}'")
print(f"Vector Dimensions: {len(sample_embedding)}")
print(f"Sample Values:     {[f'{v:.6f}' for v in sample_embedding[:5]]}")
print(f"Value Range:       [{min(sample_embedding):.6f}, {max(sample_embedding):.6f}]")
print("=" * 60)
````

**Why Aurora for RAG?**

Traditional RAG needs separate vector database + PostgreSQL. Aurora with pgvector unifies both:

- **Unified Storage:** Vectors + business data in one database
- **Business Logic:** Combine semantic search with price/rating/inventory filters  
- **Cost:** 60-75% savings vs separate vector DB
- **Performance:** Sub-100ms queries with HNSW indexes

**Example combining semantic + business logic:**
```sql
-- Find "running headphones" under $150, in stock, 4+ stars
SELECT name, price, rating 
FROM product_catalog
WHERE embedding <=> query_vector < 0.3    -- Semantic
  AND price <= 150                        -- Business filter
  AND availability_status = 'In Stock'    -- Business filter
  AND rating >= 4.0                       -- Business filter
LIMIT 5;
```

**Performance:** <100ms on 21,704 products thanks to HNSW indexing.

### Understanding Vector Similarity

**Cosine distance (`<=>`)** measures how similar two vectors are—the mathematical foundation of semantic search.

**Distance interpretation:**

| Distance | Meaning | Use Case |
|----------|---------|----------|
| 0.0 - 0.3 | Highly similar | Good matches (we use 0.3) |
| 0.3 - 0.5 | Somewhat related | Broader search |
| 0.5 - 1.0 | Loosely related | Usually filtered out |
| 1.0+ | Different concepts | Not relevant |

**Why 0.3?** Balances precision (finding good matches) with recall (not missing relevant products). Adjust based on your use case.

````python
def basic_semantic_search(query_text: str, limit: int = 5) -> List[Dict]:
    """
    Execute semantic similarity search using vector embeddings and cosine distance.
    
    Pipeline:
        1. Generate embedding vector from natural language query
        2. Compute cosine distance between query vector and all product embeddings
        3. Return top-k products ranked by semantic similarity (ascending distance)
    
    Args:
        query_text: Natural language search query
        limit: Maximum number of results to return (default: 5)
        
    Returns:
        List of product dictionaries with similarity metrics
    """
    # Generate query embedding vector
    query_embedding = generate_embedding(query_text)
    
    with psycopg.connect(conn_string) as conn:
        with conn.cursor(row_factory=dict_row) as cur:
            # Cosine distance operator (<=>): smaller values = higher semantic similarity
            # Results ordered by ascending distance (most similar products first)
            cur.execute("""
                SELECT 
                    "productId",
                    product_description,
                    price,
                    stars,
                    reviews,
                    "category_name" as category,
                    embedding <=> %s::vector AS distance,
                    1 - (embedding <=> %s::vector) AS similarity
                FROM bedrock_integration.product_catalog
                WHERE embedding IS NOT NULL
                ORDER BY distance ASC
                LIMIT %s
            """, (query_embedding, query_embedding, limit))
            
            return [dict(row) for row in cur.fetchall()]


# ============================================================================
# Semantic Search Demonstration
# ============================================================================

search_query = "something to keep coffee hot"
result_count = 3

print(f"🔍 Query: '{search_query}'")
print("=" * 80)

results = basic_semantic_search(search_query, limit=result_count)

for rank, product in enumerate(results, 1):
    # Extract product details
    description = product['product_description']
    price = product['price']
    rating = product['stars']
    review_count = product['reviews']
    similarity_score = product['similarity']
    
    # Format output with professional structure
    print(f"\nRank {rank}:")
    print(f"  Product: {description[:70]}{'...' if len(description) > 70 else ''}")
    print(f"  Price: ${price:.2f} | Rating: {rating}⭐ ({review_count:,} reviews)")
    print(f"  Similarity Score: {similarity_score:.4f}")

print("\n" + "=" * 80)
print(f"✅ Retrieved {len(results)} semantically similar products")
````

## 2D. HNSW Index Tuning - Performance Optimization

**HNSW index parameters** control the speed vs accuracy trade-off:

| Parameter | What it does | Workshop value | Production range |
|-----------|--------------|----------------|------------------|
| `m` | Graph connections per layer | 16 | 12-48 |
| `ef_construction` | Build-time search depth | 128 | 100-400 |
| `ef_search` | Query-time search depth | 40 (default) | 10-200 |

**Key insight:** `ef_search` is the only **runtime parameter**—you can tune it per-query without rebuilding the index.

### The Trade-off

**Lower ef_search (10-30):**
- ⚡ Faster queries (2-3x speedup)
- Use for: High volume, autocomplete, real-time features

**Higher ef_search (80-160):**
- 🎯 Better accuracy (closer to exact)
- Use for: Critical decisions, low volume, top-5 results

**Our setup:** `m=16, ef_construction=128, ef_search=40` → Balanced for e-commerce (~95% recall, <50ms queries)

### Production Guidelines

| Use Case | ef_search | Why |
|----------|-----------|-----|
| E-commerce search | 20-40 | High volume, good enough accuracy |
| Autocomplete | 10-20 | Speed critical, approximate OK |
| Financial/Medical | 80-160 | Accuracy critical, lower volume |

💡 **Pro tip:** Start with `ef_search=40`. Increase only if you need better recall. Lower for speed gains.

<details>
<summary style="cursor: pointer; color: #c4b5fd; font-weight: 600; padding: 8px 0;">
📚 Optional: What is HNSW? (Click to expand)
</summary>

**HNSW (Hierarchical Navigable Small World)** is a graph-based algorithm for fast approximate search.

**How it works:** Instead of checking all 21,704 products (~19,000 memory pages), HNSW builds a multi-layer graph that navigates to similar vectors by traversing ~40 connections.

**Analogy:** Like using highways to navigate a city—start on interstate (top layer), exit to main roads (middle), then local streets (bottom). Check ~40 locations instead of every house.

**Result:** 99.8% memory reduction (40 pages vs 19,000), sub-50ms queries.
