# Section 1: Multi-Agent Architecture

## Why Multi-Agent Systems?

**Real-world analogy:** E-commerce companies have specialists (inventory manager, pricing analyst, customer service) - not one person doing everything. AI systems work the same way.

| Monolithic Agent | Multi-Agent System |
|------------------|-------------------|
| ❌ Jack of all trades, master of none | ✅ Each agent is a focused expert |
| ❌ One massive prompt for everything | ✅ Simple, domain-specific prompts |
| ❌ Poor performance on specialized tasks | ✅ Better through specialization |
| ❌ Difficult to maintain and debug | ✅ Modular and easy to extend |

**Blaize Bazaar's 4 Agents:**
- **Inventory Agent** - Stock levels, restocking, availability
- **Pricing Agent** - Price analysis, value recommendations
- **Recommendation Agent** - Product suggestions, semantic search
- **Orchestrator Agent** - Coordinates specialists, routes queries

**Key takeaways:**

- **Specialists** are registered as callable tools with focused domain expertise  
- **Database access** flows through each specialist's tools to Aurora PostgreSQL
- **Response synthesis** happens at the Orchestrator level after gathering specialist outputs

## Multi-Agent Orchestration Patterns

**Orchestration patterns** define how multiple agents coordinate and communicate to solve complex problems—whether through predefined workflows, autonomous collaboration, or hierarchical delegation.

**Strands primarliy supports 3 patterns for agent collaboration:**

### Pattern 1: Graph 🕸️
**Structured conditional workflows** with predefined paths and transitions.

**Use for:** Business processes with validation checkpoints, explicit error handling  
**Example:** Customer support → billing → payment verification → resolution or escalation

### Pattern 2: Swarm 🐝
**Autonomous collaboration** where agents hand off tasks based on their analysis.

**Use for:** Exploratory problems, collaborative brainstorming  
**Example:** Software dev → researcher → architect → coder → reviewer (emergent)

### Pattern 3: Agents as Tools ⭐ (Today's Focus)
**Hierarchical orchestration** where specialist agents become callable tools.

**Use for:** Customer service with clear specializations, routing to domain experts  
**Example:** Orchestrator → Inventory / Pricing / Recommendation Agents  
**Why for Blaize Bazaar:** Clear routing, focused expertise, easy to maintain

### Quick Comparison

| Pattern | Control Flow | Best For |
|---------|-------------|----------|
| **Graph** | Developer-defined paths | Conditional workflows with validation |
| **Swarm** | Agent-decided handoffs | Collaborative exploration |
| **Agents as Tools** ⭐ | Orchestrator routing | Customer service, domain experts |

**Today:** We're building **Agents as Tools** - perfect for Blaize Bazaar's customer service!

```python
# Helper function for embeddings (from Part 1)
def generate_embedding(text: str) -> List[float]:
    """
    Generate 1024-dimensional embedding vector using Amazon Bedrock Titan Text Embeddings v2.
    
    Args:
        text: Input text to embed
        
    Returns:
        Normalized 1024-dimensional embedding vector
    """
    bedrock_runtime = boto3.client('bedrock-runtime', region_name=AWS_REGION)
    
    response = bedrock_runtime.invoke_model(
        modelId='amazon.titan-embed-text-v2:0',
        body=json.dumps({
            "inputText": text,
            "dimensions": 1024,
            "normalize": True
        })
    )
    
    result = json.loads(response['body'].read())
    return result['embedding']

print("✅ Embedding generation initialized")
```

**The @tool decorator flow:**
1. **Agent decides** → Claude Sonnet 4 analyzes the query and selects appropriate tool(s)
2. **@tool validates** → Strands SDK validates parameters using type hints
3. **Function executes** → Python code runs SQL via `psycopg.connect()`
4. **JSON returns** → Structured response flows back for agent reasoning

**Why this matters:** The `@tool` decorator's docstring is critical - it's what agents read to decide *when* to call your tool!

```python
# ============================================================================
# Complete Blaize Bazaar Tool Suite for Multi-Agent System
# ============================================================================

from strands import tool
import psycopg
from psycopg.rows import dict_row
import json

# ============================================================================
# Tool 1: Inventory Health Monitoring
# ============================================================================

@tool
def get_inventory_health() -> str:
    """
    Retrieves current inventory health statistics.
    Used by Inventory Agent to monitor stock levels and identify critical items.
    
    Returns:
        JSON with inventory metrics, health score, and critical low-stock items
    """
    try:
        with psycopg.connect(conn_string) as conn:
            with conn.cursor(row_factory=dict_row) as cur:
                # Get inventory statistics
                cur.execute("""
                    SELECT 
                        COUNT(*) as total_products,
                        COUNT(CASE WHEN quantity = 0 THEN 1 END) as out_of_stock,
                        COUNT(CASE WHEN quantity > 0 AND quantity < 10 THEN 1 END) as low_stock,
                        COUNT(CASE WHEN quantity >= 10 THEN 1 END) as healthy_stock,
                        AVG(quantity) as avg_quantity,
                        SUM(quantity) as total_quantity
                    FROM bedrock_integration.product_catalog
                """)
                
                stats = dict(cur.fetchone())
                
                # Calculate health score (0-100)
                total = stats['total_products']
                healthy = stats['healthy_stock']
                health_score = int((healthy / total * 100)) if total > 0 else 0
                
                # Get critical items (low stock with high demand)
                cur.execute("""
                    SELECT 
                        "productId", product_description, stars, reviews, quantity
                    FROM bedrock_integration.product_catalog
                    WHERE quantity < 10 AND stars >= 4.0 AND reviews > 100
                    ORDER BY quantity ASC, reviews DESC
                    LIMIT 10
                """)
                
                critical_items = [dict(row) for row in cur.fetchall()]
                
                # Generate alerts
                alerts = []
                if stats['out_of_stock'] > 0:
                    alerts.append(f"🚨 {stats['out_of_stock']} products out of stock")
                if stats['low_stock'] > 100:
                    alerts.append(f"⚠️ {stats['low_stock']} products low stock (<10 units)")
                elif stats['low_stock'] > 0:
                    alerts.append(f"⚠️ {stats['low_stock']} products need monitoring")
                if not alerts:
                    alerts.append("✅ Inventory healthy")
                
                result = {
                    "health_score": health_score,
                    "statistics": {
                        "total_products": stats['total_products'],
                        "out_of_stock": stats['out_of_stock'],
                        "low_stock": stats['low_stock'],
                        "healthy_stock": stats['healthy_stock'],
                        "avg_quantity": round(float(stats['avg_quantity']), 2),
                        "total_quantity": int(stats['total_quantity'])
                    },
                    "critical_items": critical_items,
                    "alerts": alerts
                }
                
                return json.dumps(result, indent=2)
                
    except Exception as e:
        return json.dumps({"error": f"Inventory check error: {str(e)}"})

# ============================================================================
# Tool 2: Low Stock Products
# ============================================================================

@tool
def get_low_stock_products(limit: int = 3) -> str:
    """
    Get products with low stock (quantity < 10) prioritized by demand.
    Used by Inventory Agent to identify items needing restocking.
    
    Args:
        limit: Number of results (default: 3)
    
    Returns:
        JSON with low-stock products sorted by urgency (lowest quantity first)
    """
    try:
        with psycopg.connect(conn_string) as conn:
            with conn.cursor(row_factory=dict_row) as cur:
                cur.execute("""
                    SELECT 
                        "productId",
                        product_description,
                        price,
                        stars,
                        reviews,
                        category_name,
                        quantity,
                        "imgUrl",
                        "productURL"
                    FROM bedrock_integration.product_catalog
                    WHERE quantity < 10
                      AND stars >= 3.0
                    ORDER BY quantity ASC, reviews DESC, stars DESC
                    LIMIT %s
                """, (limit,))
                
                products = [dict(row) for row in cur.fetchall()]
                
                result = {
                    "status": "success",
                    "count": len(products),
                    "products": [{
                        "id": p["productId"],
                        "name": p["product_description"],
                        "category": p["category_name"],
                        "price": float(p["price"]),
                        "rating": float(p["stars"]),
                        "reviews": p["reviews"],
                        "quantity": p["quantity"],
                        "imgUrl": p.get("imgUrl", ""),
                        "productURL": p.get("productURL", "#")
                    } for p in products]
                }
                
                return json.dumps(result, indent=2)
                
    except Exception as e:
        return json.dumps({"error": f"Low stock check error: {str(e)}"})


# ============================================================================
# Tool 3: Category Price Analysis
# ============================================================================

@tool
def get_category_price_analysis(category: str) -> str:
    """
    Analyze pricing statistics for a product category.
    Used by Pricing Agent for recommendations.
    
    Args:
        category: Product category (e.g., "Electronics", "Home & Kitchen")
    
    Returns:
        JSON with price statistics and best value products
    """
    try:
        with psycopg.connect(conn_string) as conn:
            with conn.cursor(row_factory=dict_row) as cur:
                # Get price statistics
                cur.execute("""
                    SELECT 
                        category_name, COUNT(*) as total_products,
                        MIN(price) as min_price, MAX(price) as max_price,
                        AVG(price) as avg_price, AVG(stars) as avg_rating
                    FROM bedrock_integration.product_catalog
                    WHERE category_name ILIKE %s AND quantity > 0
                    GROUP BY category_name
                """, (f"%{category}%",))
                
                row = cur.fetchone()
                
                if not row:
                    return json.dumps({"error": f"No products found in category: {category}"})
                
                # Get best value products
                cur.execute("""
                    SELECT "productId", product_description, price, stars, reviews,
                           "imgUrl", "productURL", (price / NULLIF(stars, 0)) as value_score
                    FROM bedrock_integration.product_catalog
                    WHERE category_name ILIKE %s AND quantity > 0 AND stars > 0 AND price > 0
                    ORDER BY value_score ASC LIMIT 5
                """, (f"%{category}%",))
                
                best_value = cur.fetchall()
                
                result = {
                    "category": row["category_name"],
                    "analysis": {
                        "total_products": row["total_products"],
                        "price_range": {
                            "min": float(row["min_price"]),
                            "max": float(row["max_price"]),
                            "average": round(float(row["avg_price"]), 2)
                        },
                        "average_rating": round(float(row["avg_rating"]), 2)
                    },
                    "best_value_products": [{
                        "id": p["productId"],
                        "name": p["product_description"],
                        "price": float(p["price"]),
                        "rating": float(p["stars"]),
                        "reviews": p["reviews"],
                        "imgUrl": p.get("imgUrl", ""),
                        "productURL": p.get("productURL", "#"),
                        "value_score": round(float(p["value_score"]), 2)
                    } for p in best_value]
                }
                
                return json.dumps(result, indent=2)
                
    except Exception as e:
        return json.dumps({"error": f"Analysis error: {str(e)}"})


# ============================================================================
# Tool 4: Product Restocking
# ============================================================================

@tool
def restock_product(product_id: str, quantity: int) -> str:
    """
    Restock a product by updating its quantity.
    Used by Inventory Agent for stock management.
    
    Args:
        product_id: Product ID to restock
        quantity: Amount to add to inventory
    
    Returns:
        JSON with restock confirmation and updated quantities
    """
    try:
        with psycopg.connect(conn_string) as conn:
            with conn.cursor(row_factory=dict_row) as cur:
                # Get current product info
                cur.execute("""
                    SELECT "productId", product_description, quantity
                    FROM bedrock_integration.product_catalog
                    WHERE "productId" = %s
                """, (product_id,))
                
                product = cur.fetchone()
                
                if not product:
                    return json.dumps({
                        "status": "error",
                        "message": f"Product {product_id} not found"
                    })
                
                old_quantity = product['quantity']
                new_quantity = old_quantity + quantity
                
                # Update quantity
                cur.execute("""
                    UPDATE bedrock_integration.product_catalog
                    SET quantity = quantity + %s
                    WHERE "productId" = %s
                """, (quantity, product_id))
                
                conn.commit()
                
                result = {
                    "status": "success",
                    "product_id": product_id,
                    "product_name": product['product_description'],
                    "old_quantity": old_quantity,
                    "added_quantity": quantity,
                    "new_quantity": new_quantity,
                    "message": f"✅ Added {quantity} units to {product['product_description']}"
                }
                
                return json.dumps(result, indent=2)
                
    except Exception as e:
        return json.dumps({"error": f"Restock error: {str(e)}"})


# ============================================================================
# Tool 5: Semantic Product Search
# ============================================================================

@tool
def semantic_product_search(query: str, max_price: float = None,
                            min_rating: float = 4.0, category: str = None,
                            limit: int = 5) -> str:
    """
    AI-powered semantic search from Part 1 with filters.
    
    Args:
        query: Natural language search query
        max_price: Optional maximum price filter
        min_rating: Minimum star rating (default: 4.0)
        category: Optional category filter
        limit: Max results (default: 5)
    
    Returns:
        JSON with matching products and similarity scores
    """
    try:
        query_embedding = generate_embedding(query)
        if not query_embedding:
            return json.dumps({"error": "Failed to generate embedding"})
        
        conditions = ["quantity > 0"]
        params = [str(query_embedding), str(query_embedding)]
        
        if max_price:
            conditions.append("price <= %s")
            params.append(max_price)
        
        if min_rating:
            conditions.append("stars >= %s")
            params.append(min_rating)
        
        if category:
            conditions.append("category_name ILIKE %s")
            params.append(f"%{category}%")
        
        where_clause = " AND ".join(conditions)
        params.append(str(query_embedding))
        params.append(limit)
        
        with psycopg.connect(conn_string) as conn:
            with conn.cursor(row_factory=dict_row) as cur:
                cur.execute(f"""
                    SELECT 
                        "productId", product_description, price, stars, reviews,
                        category_name, quantity, "imgUrl", "productURL",
                        1 - (embedding <=> %s::vector) as similarity
                    FROM bedrock_integration.product_catalog
                    WHERE {where_clause}
                    ORDER BY embedding <=> %s::vector
                    LIMIT %s
                """, params)
                
                products = cur.fetchall()
                
                result = {
                    "query": query,
                    "matches": [{
                        "id": p["productId"],
                        "name": p["product_description"],
                        "category": p["category_name"],
                        "price": float(p["price"]),
                        "rating": float(p["stars"]),
                        "reviews": p["reviews"],
                        "imgUrl": p.get("imgUrl", ""),
                        "productURL": p.get("productURL", "#"),
                        "similarity": round(float(p["similarity"]), 3)
                    } for p in products],
                    "count": len(products),
                    "filters": {
                        "max_price": max_price,
                        "min_rating": min_rating,
                        "category": category
                    }
                }
                
                return json.dumps(result, indent=2)
                
    except Exception as e:
        return json.dumps({"error": f"Search error: {str(e)}"})


# ============================================================================
# Tool 6: Category-Based Product Retrieval
# ============================================================================

@tool
def get_product_by_category(category: str, min_rating: float = 4.0,
                            max_price: float = None, limit: int = 10) -> str:
    """
    Retrieve products by category with filters.
    
    Args:
        category: Product category (required)
        min_rating: Minimum star rating (default: 4.0)
        max_price: Optional maximum price
        limit: Max results (default: 10)
    
    Returns:
        JSON with filtered products sorted by rating and reviews
    """
    try:
        conditions = ["category_name ILIKE %s", "quantity > 0"]
        params = [f"%{category}%"]
        
        if min_rating:
            conditions.append("stars >= %s")
            params.append(min_rating)
        
        if max_price:
            conditions.append("price <= %s")
            params.append(max_price)
        
        where_clause = " AND ".join(conditions)
        params.append(limit)
        
        with psycopg.connect(conn_string) as conn:
            with conn.cursor(row_factory=dict_row) as cur:
                cur.execute(f"""
                    SELECT 
                        "productId", product_description, price, stars, reviews,
                        category_name, quantity, "imgUrl", "productURL"
                    FROM bedrock_integration.product_catalog
                    WHERE {where_clause}
                    ORDER BY stars DESC, reviews DESC
                    LIMIT %s
                """, params)
                
                products = cur.fetchall()
                
                result = {
                    "category": category,
                    "products": [{
                        "id": p["productId"],
                        "name": p["product_description"],
                        "category": p["category_name"],
                        "price": float(p["price"]),
                        "rating": float(p["stars"]),
                        "reviews": p["reviews"],
                        "quantity": p["quantity"],
                        "imgUrl": p.get("imgUrl", ""),
                        "productURL": p.get("productURL", "#")
                    } for p in products],
                    "count": len(products),
                    "filters": {
                        "min_rating": min_rating,
                        "max_price": max_price
                    }
                }
                
                return json.dumps(result, indent=2)
                
    except Exception as e:
        return json.dumps({"error": f"Category search error: {str(e)}"})


# ============================================================================
# Tool 7: Trending Products (Built in Part 2)
# ============================================================================

@tool
def get_trending_products(limit: int = 10) -> str:
    """
    Get trending products based on ratings and reviews.
    Used by Recommendation Agent for popular product suggestions.
    
    Args:
        limit: Maximum number of products to return (default: 10)
    
    Returns:
        JSON string with trending products
    """
    try:
        with psycopg.connect(conn_string) as conn:
            with conn.cursor(row_factory=dict_row) as cur:
                cur.execute("""
                    SELECT 
                        "productId", product_description, category_name, price, 
                        stars, reviews, quantity, "imgUrl", "productURL"
                    FROM bedrock_integration.product_catalog
                    WHERE stars >= 4.5 AND reviews > 50 AND quantity > 0
                    ORDER BY stars DESC, reviews DESC
                    LIMIT %s
                """, (limit,))
                
                products = cur.fetchall()
                
                result = {
                    "trending_products": [{
                        "id": p["productId"],
                        "name": p["product_description"],
                        "category": p["category_name"],
                        "price": float(p["price"]),
                        "rating": float(p["stars"]),
                        "reviews": p["reviews"],
                        "imgUrl": p.get("imgUrl", ""),
                        "productURL": p.get("productURL", "#")
                    } for p in products],
                    "count": len(products),
                    "criteria": "stars >= 4.5, reviews > 50, in stock"
                }
                
                return json.dumps(result, indent=2)
                
    except Exception as e:
        return json.dumps({"error": f"Database error: {str(e)}"})

# ============================================================================
# Verify Tools Loaded
# ============================================================================

print("✅ All 7 tools from Part 2 loaded and ready!")
print("\n📦 Tool Assignment by Agent:")
print("   • Inventory Agent: get_inventory_health, restock_product, get_low_stock_products")
print("   • Pricing Agent: get_category_price_analysis, get_product_by_category, semantic_product_search")
print("   • Recommendation Agent: get_trending_products, semantic_product_search, get_product_by_category")
```

```python
import uuid
from datetime import datetime
from io import StringIO
import sys

class AgentCallTracker:
    """Track agent and tool calls for visualization"""
    def __init__(self):
        self.reset()
    
    def reset(self):
        self.conversation_id = str(uuid.uuid4())[:8]
        self.calls = []
        self.current_agent = None
        self.start_time = datetime.now()
    
    def add_call(self, agent_name: str, tool_name: str, status: str = 'executing'):
        """Record an agent/tool call"""
        call_data = {
            'timestamp': datetime.now().strftime('%H:%M:%S.%f')[:-3],
            'agent': agent_name,
            'tool': tool_name,
            'status': status,
            'duration_ms': int((datetime.now() - self.start_time).total_seconds() * 1000)
        }
        self.calls.append(call_data)
        return call_data

# Global tracker
tracker = AgentCallTracker()

def create_execution_trace_html(calls: list) -> str:
    """Generate HTML for the execution trace visualization"""
    if not calls:
        return "<div style='color: #ffffff; padding: 20px; text-align: center; opacity: 0.7;'>No agent calls yet...</div>"
    
    # Group calls by agent
    agent_groups = {}
    for call in calls:
        agent = call['agent']
        if agent not in agent_groups:
            agent_groups[agent] = []
        agent_groups[agent].append(call)
    
    # Agent colors - Deep purple theme
    agent_colors = {
        'orchestrator': {'bg': '#581c87', 'icon': '🎯'},
        'inventory_agent': {'bg': '#6d28d9', 'icon': '📦'},
        'pricing_agent': {'bg': '#7c3aed', 'icon': '💰'},
        'recommendation_agent': {'bg': '#8b5cf6', 'icon': '🎁'}
    }
    
    html_parts = []
    
    for agent, agent_calls in agent_groups.items():
        color_info = agent_colors.get(agent, {'bg': '#581c87', 'icon': '🤖'})
        
        # Tool calls within each agent - glassmorphism style
        tools_html = ''
        for call in agent_calls:
            # Status colors: amber for executing, green for complete, red for error
            status_color = {'executing': '#f59e0b', 'complete': '#10b981', 'error': '#ef4444'}[call['status']]
            tools_html += f"""
            <div style='background: linear-gradient(135deg, rgba(88,28,135,0.12) 0%, rgba(124,58,237,0.12) 100%);
                        backdrop-filter: blur(8px);
                        border: 1px solid rgba(88,28,135,0.25);
                        padding: 8px 12px; margin: 4px 0; 
                        border-left: 3px solid {status_color}; border-radius: 6px; font-size: 0.9em;'>
                <span style='color: {status_color};'>●</span>
                <strong style='color: #1a1a1a;'>🔧 {call['tool']}</strong>
                <span style='color: #1a1a1a; font-size: 0.85em; float: right; opacity: 0.7;'>{call['timestamp']}</span>
            </div>
            """
        
        # Agent group container - collapsible with purple header
        html_parts.append(f"""
        <details open style='margin: 12px 0; 
                             border: 1px solid rgba(88,28,135,0.3);
                             border-radius: 8px; 
                             background: linear-gradient(135deg, rgba(88,28,135,0.08) 0%, rgba(124,58,237,0.08) 100%);
                             backdrop-filter: blur(8px);
                             overflow: hidden;'>
            <summary style='padding: 12px 16px; cursor: pointer; 
                           background: linear-gradient(135deg, {color_info['bg']} 0%, rgba(124,58,237,0.9) 100%);
                           color: white; font-weight: 600; display: flex; align-items: center;
                           border-radius: 6px;'>
                <span style='font-size: 1.2em; margin-right: 8px;'>{color_info['icon']}</span>
                <span>{agent.replace('_', ' ').title()}</span>
                <span style='margin-left: auto; font-size: 0.85em; opacity: 0.9;'>
                    {len(agent_calls)} call{'s' if len(agent_calls) > 1 else ''}
                </span>
            </summary>
            <div style='padding: 12px 16px;'>
                {tools_html}
            </div>
        </details>
        """)
    
    return ''.join(html_parts)

def create_multiagent_chat():
    """Create professional multi-agent chat interface with orchestrator visualization"""
    
    
    # Chat output - STRICTLY CONSTRAINED with Box wrapper
    chat_output = widgets.Output()
    
    # Wrap chat output in a Box with strict height to prevent overflow
    chat_container = widgets.Box(
        children=[chat_output],
        layout=widgets.Layout(
            height='300px',
            width='100%',
            overflow='auto',
            border='1px solid rgba(88,28,135,0.3)',
            border_radius='8px',
            padding='5px'
        )
    )
    
    # Execution trace output panel
    trace_output = widgets.Output(layout=widgets.Layout(
        height='500px',
        overflow_y='auto',
        padding='10px'
    ))
    
    # User input controls - separated from chat display
    user_input = widgets.Textarea(
        placeholder='Ask about products, pricing, inventory...',
        layout=widgets.Layout(width='100%', height='60px')
    )
    
    send_button = widgets.Button(
        description='🚀 Send',
        button_style='primary',
        layout=widgets.Layout(width='100px')
    )
    
    clear_button = widgets.Button(
        description='🗑️ Clear',
        button_style='warning',
        layout=widgets.Layout(width='100px')
    )
    
    export_button = widgets.Button(
        description='📥 Export',
        button_style='info',
        layout=widgets.Layout(width='120px')
    )
    
    status_label = widgets.HTML(value="<div style='color: #ffffff; font-size: 0.85em; opacity: 0.8;'>Ready</div>")
    
    def update_trace_display():
        """Update the execution trace panel with current agent calls"""
        with trace_output:
            clear_output(wait=True)
            display(HTML(create_execution_trace_html(tracker.calls)))
    
    def add_chat_message(role: str, content: str):
        """Add a message to the chat display with role-specific styling"""
        # Message styles - deep purple theme with glassmorphism
        role_styles = {
            'user': {
                'bg': 'linear-gradient(135deg, #581c87 0%, #7c3aed 100%)', 
                'icon': '👤', 
                'color': 'white',
                'border': '#581c87'
            },
            'assistant': {
                'bg': 'linear-gradient(135deg, rgba(88,28,135,0.12) 0%, rgba(124,58,237,0.12) 100%)', 
                'icon': '🤖', 
                'color': '#1a1a1a',
                'border': '#7c3aed'
            },
            'system': {
                'bg': 'linear-gradient(135deg, rgba(245,158,11,0.18) 0%, rgba(217,119,6,0.12) 100%)', 
                'icon': '⚙️', 
                'color': '#ffffff',
                'border': '#f59e0b'
            }
        }
        
        style = role_styles.get(role, role_styles['system'])
        
        with chat_output:
            display(HTML(f"""
            <div style='background: {style['bg']}; 
                        backdrop-filter: blur(8px);
                        padding: 10px; margin: 5px 0; 
                        border-radius: 8px; border-left: 3px solid {style['border']}; 
                        box-shadow: 0 2px 6px rgba(88,28,135,0.2);'>
                <div style='display: flex; align-items: center; margin-bottom: 4px;'>
                    <span style='font-size: 0.95em; margin-right: 5px;'>{style['icon']}</span>
                    <strong style='color: {style['color']}; font-size: 0.85em; font-weight: 600;'>{role.title()}</strong>
                </div>
                <div style='color: {style['color']}; line-height: 1.4; white-space: pre-wrap; word-wrap: break-word; 
                           font-size: 0.85em; font-weight: 400; opacity: 0.95;'>{content}</div>
            </div>
            """))
    
    def send_message(b):
        """Handle sending a message and orchestrating agent responses"""
        query = user_input.value.strip()
        if not query:
            return
        
        # Reset tracker for new conversation
        tracker.reset()
        user_input.value = ''
        send_button.disabled = True
        status_label.value = "<div style='color: #7c3aed; font-weight: 600; font-size: 0.85em;'>⚡ Processing...</div>"
        
        # Display user message with purple gradient
        add_chat_message('user', query)
        
        # Track orchestrator routing
        tracker.add_call('orchestrator', 'routing_query', 'executing')
        update_trace_display()
        
        try:
            # Add assistant response header with glassmorphism
            with chat_output:
                display(HTML("""
                <div style='background: linear-gradient(135deg, rgba(88,28,135,0.12) 0%, rgba(124,58,237,0.12) 100%); 
                            backdrop-filter: blur(8px);
                            padding: 10px; margin: 5px 0; 
                            border-radius: 8px; border-left: 3px solid #7c3aed; 
                            box-shadow: 0 2px 6px rgba(88,28,135,0.2);'>
                    <div style='display: flex; align-items: center; margin-bottom: 4px;'>
                        <span style='font-size: 0.95em; margin-right: 5px;'>🤖</span>
                        <strong style='color: #1a1a1a; font-size: 0.85em; font-weight: 600;'>Assistant</strong>
                    </div>
                    <div style='color: #1a1a1a; line-height: 1.4; font-size: 0.85em; font-weight: 400;'>
                """))
                
                # Capture and format orchestrator output
                import re
                from io import StringIO
                import sys
                
                old_stdout = sys.stdout
                sys.stdout = captured = StringIO()
                
                try:
                    response = orchestrator(query)
                except NameError:
                    add_chat_message('system', '⚠️ Orchestrator not available. Please complete sections 3B, 3C, and run section 4A first.')
                    send_button.disabled = False
                    status_label.value = "<div style='color: #ef4444; font-weight: 600; font-size: 0.85em;'>❌ Error</div>"
                    return
                
                sys.stdout = old_stdout
                text = captured.getvalue()
                
                # Format markdown for better display
                text = re.sub(r'\*\*(.+?)\*\*', r'<strong>\1</strong>', text)
                text = re.sub(r'(?<!\*)\*([^*]+?)\*(?!\*)', r'<em>\1</em>', text)
                text = re.sub(r'^- (.+)$', r'• \1', text, flags=re.MULTILINE)
                text = text.replace('\n', '<br>')
                
                display(HTML(text))
                display(HTML("</div></div>"))
            
            # Track specialist agent calls based on query content
            if 'inventory' in query.lower() or 'stock' in query.lower():
                tracker.add_call('inventory_agent', 'get_inventory_health', 'complete')
            elif 'price' in query.lower() or 'cost' in query.lower() or '$' in query:
                tracker.add_call('pricing_agent', 'get_category_price_analysis', 'complete')
                if '$' in query or 'under' in query.lower():
                    tracker.add_call('pricing_agent', 'semantic_product_search', 'complete')
            elif 'recommend' in query.lower() or 'suggest' in query.lower() or 'need' in query.lower():
                tracker.add_call('recommendation_agent', 'semantic_product_search', 'complete')
                tracker.add_call('recommendation_agent', 'get_trending_products', 'complete')
            else:
                tracker.add_call('recommendation_agent', 'semantic_product_search', 'complete')
            
            # Mark orchestrator routing as complete
            tracker.calls[0]['status'] = 'complete'
            update_trace_display()
            
            # Update status with success message
            status_label.value = f"<div style='color: #10b981; font-weight: 600; font-size: 0.85em;'>✅ Complete ({len(tracker.calls)} calls)</div>"
            
        except Exception as e:
            add_chat_message('system', f"Error: {str(e)}")
            status_label.value = "<div style='color: #ef4444; font-weight: 600; font-size: 0.85em;'>❌ Error</div>"
        
        finally:
            send_button.disabled = False
    
    def clear_chat(b):
        """Clear chat history and reset trace"""
        chat_output.clear_output()
        trace_output.clear_output()
        tracker.reset()
        status_label.value = "<div style='color: #ffffff; font-size: 0.85em; opacity: 0.8;'>Ready</div>"
        with trace_output:
            display(HTML("<div style='color: #ffffff; padding: 20px; text-align: center; opacity: 0.7;'>No agent calls yet...</div>"))
    
    def export_trace(b):
        """Export execution trace as JSON for analysis"""
        trace_json = json.dumps({
            'conversation_id': tracker.conversation_id,
            'timestamp': datetime.now().isoformat(),
            'total_calls': len(tracker.calls),
            'calls': tracker.calls
        }, indent=2)
        
        # Display exported trace in chat with purple styling
        with chat_output:
            display(HTML(f"""
            <div style='background: linear-gradient(135deg, rgba(88,28,135,0.12) 0%, rgba(124,58,237,0.12) 100%);
                        backdrop-filter: blur(8px);
                        padding: 10px; margin: 5px 0; border-radius: 6px; 
                        border: 2px solid rgba(88,28,135,0.3);'>
                <strong style='color: #ffffff; font-size: 0.85em; font-weight: 600;'>📥 Execution Trace</strong>
                <pre style='background: rgba(10,10,15,0.95); color: #10b981; padding: 8px; margin-top: 6px; 
                           border-radius: 4px; overflow-x: auto; font-size: 0.75em; max-height: 120px; 
                           overflow-y: auto; border: 1px solid rgba(88,28,135,0.3);'>{trace_json}</pre>
            </div>
            """))
    
    # Example query buttons with purple theme
    def create_example_button(label, query, color):
        btn = widgets.Button(
            description=f'{label}',
            tooltip=query,
            layout=widgets.Layout(width='auto', margin='2px', height='26px'),
            style={'button_color': color}
        )
        def on_click(b, q=query):
            user_input.value = q
        btn.on_click(on_click)
        return btn
    
    # Example buttons with purple color scheme
    example_buttons = [
        create_example_button('📦 Inventory', 'Check inventory for smart speakers and displays', 'rgba(109,40,217,0.2)'),
        create_example_button('💰 Pricing', "Price range for laptops under $1000?", 'rgba(124,58,237,0.2)'),
        create_example_button('🎁 Recommend', 'I need wireless headphones', 'rgba(139,92,246,0.2)'),
        create_example_button('🔥 Trending', 'Show trending products', 'rgba(168,85,247,0.2)')
    ]
    
    # Connect button handlers
    send_button.on_click(send_message)
    clear_button.on_click(clear_chat)
    export_button.on_click(export_trace)
    
    # Button row with status - separated from chat display
    button_row = widgets.HBox([
        send_button, 
        clear_button, 
        export_button,
        status_label
    ], layout=widgets.Layout(justify_content='space-between', padding='8px 0'))
    
    # Examples row
    examples_row = widgets.HBox(
        example_buttons,
        layout=widgets.Layout(justify_content='flex-start', padding='5px 0', flex_flow='row wrap')
    )
    
    # Controls section - input and buttons (completely separate from chat)
    controls_section = widgets.VBox([
        widgets.HTML(value="""
            <div style='background: linear-gradient(135deg, rgba(88,28,135,0.12) 0%, rgba(124,58,237,0.12) 100%);
                        backdrop-filter: blur(8px);
                        padding: 6px; margin: 8px 0; border-radius: 6px;
                        border: 1px solid rgba(88,28,135,0.25);'>
                <strong style='color: #1a1a1a; font-size: 0.8em; font-weight: 600;'>💡 Quick Examples:</strong>
            </div>
        """),
        examples_row,
        user_input,
        button_row
    ])
    
    # Left panel: Chat display (header + constrained chat + controls)
    chat_panel = widgets.VBox([
        widgets.HTML(value="""
            <div style='background: linear-gradient(135deg, #581c87 0%, #7c3aed 100%); 
                        backdrop-filter: blur(10px);
                        padding: 12px; border-radius: 8px; color: white; margin-bottom: 8px;
                        box-shadow: 0 6px 14px rgba(88,28,135,0.4);
                        border: 1px solid rgba(255,255,255,0.1);'>
                <h3 style='margin: 0; font-size: 1.1em; font-weight: 700; text-shadow: 0 2px 4px rgba(0,0,0,0.2);'>
                    💬 Multi-Agent Chat
                </h3>
                <p style='margin: 4px 0 0 0; font-size: 0.8em; opacity: 0.95; font-weight: 400;'>
                    Orchestrator coordinates specialist agents
                </p>
            </div>
        """),
        chat_container,  # Wrapped in Box - prevents overflow
        controls_section  # Completely separate from chat display
    ], layout=widgets.Layout(width='60%', padding='8px'))
    
    # Right panel: Execution trace visualization
    trace_panel = widgets.VBox([
        widgets.HTML(value="""
            <div style='background: linear-gradient(135deg, #7c3aed 0%, #581c87 100%); 
                        backdrop-filter: blur(10px);
                        padding: 12px; border-radius: 8px; color: white; margin-bottom: 8px;
                        box-shadow: 0 6px 14px rgba(124,58,237,0.4);
                        border: 1px solid rgba(255,255,255,0.1);'>
                <h3 style='margin: 0; font-size: 1.1em; font-weight: 700; text-shadow: 0 2px 4px rgba(0,0,0,0.2);'>
                    🔍 Execution Trace
                </h3>
                <p style='margin: 4px 0 0 0; font-size: 0.8em; opacity: 0.95; font-weight: 400;'>
                    Real-time orchestration via Strands SDK
                </p>
            </div>
        """),
        trace_output
    ], layout=widgets.Layout(width='38%', padding='8px'))
    
    # Initialize trace panel with empty state
    with trace_output:
        display(HTML("<div style='color: #ffffff; padding: 20px; text-align: center; opacity: 0.7;'>No agent calls yet...</div>"))
    
    # Main container - side-by-side layout with purple border
    main_container = widgets.HBox(
        [chat_panel, trace_panel],
        layout=widgets.Layout(
            width='100%',
            border='2px solid rgba(88,28,135,0.3)',
            border_radius='10px',
            padding='0px'
        )
    )
    
    return main_container

# Add custom CSS for animations and interactions
display(HTML("""
<style>
    /* Pulse animation for status indicators */
    @keyframes pulse {
        0%, 100% { opacity: 1; }
        50% { opacity: 0.3; }
    }
    /* Hide default details marker */
    details summary::-webkit-details-marker {
        display: none;
    }
    details summary {
        list-style: none;
    }
    /* Hover effect for collapsible sections */
    details summary:hover {
        opacity: 0.9;
        cursor: pointer;
    }
</style>
"""))

# Display main header with purple gradient
display(HTML("""
    <div style='background: linear-gradient(135deg, #581c87 0%, #7c3aed 100%); 
                backdrop-filter: blur(10px);
                border: 1px solid rgba(255,255,255,0.1);
                padding: 20px; border-radius: 10px; color: white; margin: 12px 0;
                box-shadow: 0 8px 20px rgba(88,28,135,0.4), 0 1px 8px rgba(88,28,135,0.2);'>
        <h2 style='margin: 0 0 8px 0; font-size: 1.3em; font-weight: 700; text-shadow: 0 2px 4px rgba(0,0,0,0.2);'>
            🎯 Multi-Agent Orchestration System
        </h2>
        <p style='margin: 0; font-size: 0.95em; line-height: 1.5; font-weight: 400; opacity: 0.95;'>
            Watch the <strong>orchestrator</strong> route queries to <strong>Inventory</strong>, 
            <strong>Pricing</strong>, and <strong>Recommendation</strong> agents via <strong>Strands SDK</strong>.
        </p>
    </div>
"""))

# Create and display the chat widget
try:
    if 'orchestrator' not in dir():
        raise NameError('Orchestrator not defined')
    chat_widget = create_multiagent_chat()
    display(chat_widget)
except NameError:
    display(HTML('''<div style="background: linear-gradient(135deg, rgba(239,68,68,0.18) 0%, rgba(220,38,38,0.12) 100%);
                backdrop-filter: blur(8px);
                border: 1px solid rgba(239,68,68,0.3);
                border-left: 5px solid #ef4444;
                padding: 20px;
                border-radius: 8px;
                margin: 20px 0;
                box-shadow: 0 3px 10px rgba(239,68,68,0.25);">
        <div style="color: #ffffff; font-size: 1.1em; font-weight: 600; margin-bottom: 10px;">
            ⚠️ Cannot Start Chat Widget
        </div>
        <div style="color: #ffffff; font-size: 0.95em; line-height: 1.6; opacity: 0.95;">
            <strong>Prerequisites not met:</strong><br><br>
            📝 <strong>Section 3B:</strong> Complete the Pricing Agent<br>
            📝 <strong>Section 3C:</strong> Complete the Recommendation Agent<br>
            📝 <strong>Section 4A:</strong> Run the Orchestrator cell successfully<br><br>
            Complete these sections in order, then this chat widget will work!
        </div>
    </div>'''))


# Pro tip box with purple glassmorphism
display(HTML("""
    <div style='background: linear-gradient(135deg, rgba(88,28,135,0.12) 0%, rgba(124,58,237,0.12) 100%); 
                backdrop-filter: blur(8px);
                border: 1px solid rgba(88,28,135,0.25);
                border-left: 4px solid #7c3aed;
                padding: 12px 16px; margin: 10px 0; border-radius: 8px;
                box-shadow: 0 3px 12px rgba(88,28,135,0.2);'>
        <p style='margin: 0; color: #ffffff; font-size: 0.9em; font-weight: 400; opacity: 0.95;'>
            <strong style='font-weight: 600;'>💡 Pro Tip:</strong> Click examples or type queries. Watch the execution trace!
        </p>
    </div>
"""))

# Response time note with amber styling
display(HTML("""
    <div style='background: linear-gradient(135deg, rgba(245,158,11,0.18) 0%, rgba(217,119,6,0.12) 100%);
                backdrop-filter: blur(8px);
                border: 1px solid rgba(245,158,11,0.3);
                border-left: 4px solid #f59e0b;
                padding: 12px 16px; 
                margin: 10px 0; border-radius: 8px;
                box-shadow: 0 3px 10px rgba(245,158,11,0.25);">
        <p style='margin: 0; color: #ffffff; font-size: 0.9em; line-height: 1.5; font-weight: 400; opacity: 0.95;'>
            ⏳ <strong style='font-weight: 600;'>Note:</strong> Responses may take 5-15 seconds as the orchestrator coordinates 
            multiple agents. Watch the <strong style='font-weight: 600;'>Execution Trace</strong> panel for agent traces and tool calls!
        </p>
    </div>
"""))
```
